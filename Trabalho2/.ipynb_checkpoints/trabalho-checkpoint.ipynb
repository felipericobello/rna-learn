# Import

```python
import sys
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import tensorflow
import keras
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

print(f"Python version: {sys.version}")
print(f"Tensor Flow Version: {tensorflow.__version__}")
print(f"Tensorflow Keras Version: {tensorflow.keras.__version__}")
print(f"Keras Version: {keras.__version__}")
```

# Data Import

```python
dataset = tensorflow.keras.datasets.boston_housing
(train_data, train_labels), (test_data, test_labels) = dataset.load_data(seed=1)

atrib = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
dataframe = pd.DataFrame(train_data, columns=atrib)

train_data = (train_data - train_data.mean(axis=0)) / train_data.std(axis=0)
test_data  = (test_data  - test_data.mean(axis=0)) / test_data.std(axis=0)

dataframe.describe()
```

# Neural Network

```python
np.random.seed(1)
tensorflow.random.set_seed(2)

def build_model():
    model = Sequential([Dense(32, activation=tensorflow.nn.relu, input_shape=(1,)), Dense(32, activation=tensorflow.nn.relu), Dense(1)], name='rna')
    optimizer = tensorflow.keras.optimizers.RMSprop(learning_rate=0.002)
    model.compile(loss='mse', optimizer=optimizer, metrics=['mae'])
    return model
```

```python
data_shape = train_data.shape
tdata_shape = test_data.shape
train_labels = train_labels.reshape(train_labels.shape[0], 1)
test_labels = test_labels.reshape(test_labels.shape[0], 1)

historylist = []
maelist = []

class PrintDot(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs):
        if epoch % 100 == 0: print('') 
        print('.', end='')

# Treinamento da rede
for i in range(data_shape[1]):
    actual_train_data = train_data[:data_shape[0], i].reshape(data_shape[0], 1)
    model = build_model()
    history = model.fit(actual_train_data, train_labels, epochs=600, batch_size=32, validation_split=0.2, verbose=0, callbacks=[PrintDot()])

    mae = model.evaluate(test_data[:tdata_shape[0], i].reshape(tdata_shape[0], 1), test_labels, verbose=0)

    maelist.append(mae)
    historylist.append(history.history)
    keras.backend.clear_session()
```

```python
atrib = np.asarray(atrib)
for i in range(atrib.size):
    print(f"Atributo {atrib[i]} tem como erro médio absoluto: ${maelist[i][1] * 1000: .2f}.")
```
